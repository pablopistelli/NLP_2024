# Cartas de visitantes

```python
##### ACLARACIÓN #####
# Luego de las primeras pruebas el archivo cambió el enlace
# Es posible que al volver a ejecutarlo el archivo no se encuentre

# Descargar el archivo desde la URL
!wget --no-check-certificate "https://boardgamegeek.com/file/download_redirect/0c68d672ed2a2262ac3ab8eda1705bbf7a72d6b89c5d81d0/French+Visitors+cards.xlsx" -O french_visitors_cards.xlsx
```

```python
df_visitors = pd.read_excel('french_visitors_cards.xlsx', engine='openpyxl')
```

```python
df_visitors
```

En esta tabla se encuentran las cartas de visitantes de verano y luego de varias líneas vacías, las de visitantes de invierno.

Se incluirá una columna para indicar a qué estación corresponde cada visitante.

Los nombres de las cartas están en inglés y francés, y la descripción de la carta sólo en francés.

Se aplicará la traducción antes de generar el archivo para trabajar con la tabla en inglés.

```python
# Elimino las líneas vacías intermedias
df_visitors.dropna(inplace=True)

# Reseteo los índices
df_visitors.reset_index(drop=True, inplace=True)
```

```python
df_visitors
```

```python
# La fila 22 es la que indica el paso de cartas de verano a invierno
# Agrego la columna de estación
for i in range(len(df_visitors)):
  if i < 22:
    df_visitors.loc[i, 'Estación'] = 'Verano'
  if i > 22:
    df_visitors.loc[i, 'Estación'] = 'Invierno'
```

```python
df_visitors
```

```python
# Elimino la fila 22
df_visitors.drop(index=22, inplace=True)

# Vuelvo a resetear indices
df_visitors.reset_index(drop=True, inplace=True)

# Elimino la columna con nombres en francés
df_visitors.drop(columns='Nom de la carte VF', inplace=True)
```

```python
df_visitors
```

```python
# Renombro las dos primeras columnas como "Nombre" y "Efecto"
df_visitors.rename(columns={df_visitors.columns[0]: 'Nombre', df_visitors.columns[1]: 'Efecto'}, inplace=True)
```

```python
df_visitors.head()
```

```python
# Traduzco el contenido de Nombre y Efecto a español utilizando uno de los módulos de idioma del TP1
!pip install deep-translator
```

```python
from deep_translator import GoogleTranslator

for i in range(len(df_visitors)):
  df_visitors.loc[i, 'Nombre'] = GoogleTranslator(source='en', target='es').translate(df_visitors.loc[i, 'Nombre'])
  df_visitors.loc[i, 'Efecto'] = GoogleTranslator(source='fr', target='es').translate(df_visitors.loc[i, 'Efecto'])
```

```python
df_visitors
```

```python
# Guardar la tabla en un archivo de texto para continuar el procesamiento
df_visitors.to_csv(path_files+'visitors_cards.csv', index=False)
```
