```python
import re
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled
from youtube_transcript_api.formatters import TextFormatter
```

## Videos extraídos de YouTube

```python
videos_urls = [
    'https://www.youtube.com/watch?v=cFNdH88bTKs',
    'https://www.youtube.com/watch?v=UlpIzDIF2Ys',
    'https://www.youtube.com/watch?v=Pde-3rJbbHc',
    'https://www.youtube.com/watch?v=xRkt0aamraY'
]

videos_ids = []

def extract_video_id(url):
    # Expresión regular para detectar el ID del video en la URL
    pattern = r"(?:https?://)?(?:www\.)?(?:youtube\.com/watch\?v=|youtu\.be/)([a-zA-Z0-9_-]+)"
    match = re.search(pattern, url)
    if match:
        return match.group(1)
    else:
        return None

# Extracción de IDs
for url in videos_urls:
    video_id = extract_video_id(url)
    if video_id:
        videos_ids.append(video_id)
    else:
        print("No se encontró un ID de video en la URL.")
```

```python
videos_ids
```

```python
# Extracción de texto de los videos
for id in videos_ids:
    try:
        # Obtener transcripción
        transcript = YouTubeTranscriptApi.get_transcript(id, languages=['es'])

        # Formatear transcripción
        formatter = TextFormatter()
        text_transcript = formatter.format_transcript(transcript)

        # Guardar el texto en un archivo de texto para continuar el procesamiento
        with open(path_files+f"Video_{id}.txt", "w", encoding="utf-8") as f:
            f.write(text_transcript)
            print("Texto extraído y guardado en "+f"'Video_{id}.txt'.")
    except TranscriptsDisabled:
        print(f"No se pudo obtener la transcripción para el video con ID: {id}. Es posible que los subtítulos estén desactivados.")
```

# Validación de idioma y Traducción

El material recopilado está en español e inglés.

Para este trabajo elegí procesar los textos en **español**, por lo tanto antes de procesar los textos que se encuentran en inglés, serán traducidos usando las herramientas probadas en el TP1.

```python
!pip install deep-translator
!pip install langdetect
```

```python
from deep_translator import GoogleTranslator
from langdetect import detect, detect_langs

# Al intentar traducir los textos encontré una limitación de Google Translate
# respecto a la cantidad máxima de caracteres.
# Genero una función para traducir por partes:

def dividir_texto(text, max_length=4000):
    # Divido el texto
    palabras = text.split()
    # Lista de divisiones aptas para el traductor
    fragmentos = []
    # Inicializo porción de texto
    fragmento_actual = ""

    for palabra in palabras:
        # Límite alcanzado
        if len(fragmento_actual) + len(palabra) + 1 > max_length:
            fragmentos.append(fragmento_actual)
            fragmento_actual = palabra
        # Incorporación de palabra en fragmento actual
        else:
            fragmento_actual += (" " + palabra if fragmento_actual else palabra)

    # Agregar el último fragmento si no está vacío
    if fragmento_actual:
        fragmentos.append(fragmento_actual)

    return fragmentos

# Función de traducción si el idioma no es español
def translate(text):
    # Detectar el idioma
    idioma = detect(text)

    # Si el texto ya está en español lo devuelvo como está
    if idioma == 'es':
        return text

    # Si no está en español, lo divido
    fragmentos = dividir_texto(text)
    traduccion_completa = ""

    # Traduzco por partes
    for fragmento in fragmentos:
        translated = GoogleTranslator(source='auto', target='es').translate(fragmento)

        # Concatenar las las partes
        traduccion_completa += translated + " "

    return traduccion_completa.strip()
```

```python
import os

# Recorre todos los archivos de la carpeta
for filename in os.listdir(path_files):
    if filename.endswith('.txt'):
        file = os.path.join(path_files, filename)
        print(f'Procesando: {file}')
        with open(file, 'r', encoding='utf-8') as f:
          text = f.read()

        with open(file[:-4]+"_ES.txt", 'w', encoding='utf-8') as f:
          f.write(translate(text))
        print(f'Texto validado en español: {file[:-4]+"_ES.txt"}')
```
